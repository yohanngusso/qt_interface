```python
df = pd.read_csv('dataset/pythondevsurvey2017_raw_data.csv')
```

```python
df.shape
```

```python
pergunta = 'Which of the following best describes your job role(s)?'
mascara_colunas =  df.columns[df.columns.str.contains(pergunta)]
colunas = df[mascara_colunas].columns

pd.Series(
    map(lambda coluna: df[mascara_colunas][coluna].value_counts()[0], colunas),
    index=map(lambda c: c.split(':')[0], colunas)
).sort_values(ascending=False).plot(kind='barh')
```

```python
df.columns[df.columns.str.contains('country')]
```

```python
mascara_brasil = df['What country do you live in?'] == 'Brazil'
```

```python
series_brasil = pd.Series(
    map(lambda coluna: df[mascara_brasil][coluna].value_counts()[0], colunas),
    index=map(lambda c: c.split(':')[0], colunas)
)
```

```python
pd.Series(
    series_brasil.values / np.sum(series_brasil.values),
    index=map(lambda c: c.split(':')[0], colunas)
).sort_values(ascending=False).plot(kind='barh')
```

```python
pd.Series(
    map(lambda coluna: df[mascara_brasil][coluna].value_counts()[0], colunas),
    index=map(lambda c: c.split(':')[0], colunas)
).sort_values(ascending=False).plot(kind='barh')
```
